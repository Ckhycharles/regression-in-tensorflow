```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def enter_name_of_dataset ():
    name =str(input("enter the name of the dataset"))
    return name
dataset = enter_name_of_dataset()
dataset = pd.read_csv(dataset)
def print_dataset():
    print(dataset)
print_dataset()


#linear regression
def check_for_missing_values():
    a = dataset.isnull().sum()
    if(a >0):
        print("there are missing values in the dataset")
        from sklearn.impute import SimpleImputer
        imputer = SimpleImputer(missing_values=np.nan ,strategy='mean')
        imputer = imputer.fit_transform(dataset)
        dataset = imputer
        return dataset
    else:
        print("there are no missing values in the dataset")
        return dataset
check_for_missing_values()
def num_of_independent():
    columns_included = int(input("enter number of columns in independent variable"))
    return columns_included
def num_of_dependent():
    columns_excluded = int(input("enter columns tu be used as dependent variable"))
    return columns_excluded

X = dataset.iloc[:, :num_of_independent()].values
Y = dataset.iloc[:, num_of_dependent()].values
print(X)
print(Y)
def describe_dataset():
    print(dataset.describe())
My_dataset = dataset.describe()
print(My_dataset)

def split_dataset_into_test_train():
    from sklearn.model_selection import train_test_split
    training_size = int(input("Enter the training percentage between 1-100%"))
    X_train,X_test,y_train,y_test = train_test_split(X,Y, train_size=training_size,random_state=0)
    return X_train,X_test,y_train,y_test
split_dataset_into_test_train()
##linear regression
def linear_regression():
    from sklearn.linear_model import LinearRegression
    regressor = linear_regression()
    X_train ,y_train =  regressor.fit(X_train,y_train)
    return
```
